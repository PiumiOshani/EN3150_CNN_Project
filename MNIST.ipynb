```python
import torch, torch.nn as nn, torch.nn.functional as F, torch.optim as optim
from torch.utils.data import random_split, DataLoader, Subset
from torchvision import datasets, transforms, models
import matplotlib.pyplot as plt
import numpy as np, pandas as pd, seaborn as sns, os
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])
mnist_train_full = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
mnist_test_full  = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

# Combine train + test then re-split 70/15/15
full_dataset = torch.utils.data.ConcatDataset([mnist_train_full, mnist_test_full])
n_total = len(full_dataset)
n_train = int(0.70 * n_total)
n_val   = int(0.15 * n_total)
n_test  = n_total - n_train - n_val
train_set, val_set, test_set = random_split(full_dataset, [n_train, n_val, n_test])

batch_size = 128
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_set, batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_set, batch_size=batch_size, shuffle=False)

# Define Custom CNN
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool  = nn.MaxPool2d(2,2)
        self.fc1   = nn.Linear(64*7*7, 128)
        self.drop  = nn.Dropout(0.5)
        self.fc2   = nn.Linear(128, 10)
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64*7*7)
        x = F.relu(self.fc1(x))
        x = self.drop(x)
        x = self.fc2(x)
        return x

# Training utilities
def train_model(model, optimizer, criterion, train_loader, val_loader, epochs=20):
    history = {'train_loss':[], 'val_loss':[], 'train_acc':[], 'val_acc':[]}
    for ep in range(epochs):
        model.train()
        correct, total, running_loss = 0,0,0
        for x, y in train_loader:
            optimizer.zero_grad()
            out = model(x)
            loss = criterion(out,y)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            pred = out.argmax(1)
            correct += (pred==y).sum().item(); total += y.size(0)
        train_loss = running_loss/len(train_loader)
        train_acc  = correct/total
        # validation
        model.eval(); v_correct,v_total,v_loss=0,0,0
        with torch.no_grad():
            for x,y in val_loader:
                out = model(x)
                loss = criterion(out,y)
                v_loss += loss.item()
                pred = out.argmax(1)
                v_correct += (pred==y).sum().item(); v_total += y.size(0)
        val_loss = v_loss/len(val_loader); val_acc = v_correct/v_total
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        print(f"Epoch {ep+1:02d}: train_loss={train_loss:.4f} val_loss={val_loss:.4f} train_acc={train_acc:.4f} val_acc={val_acc:.4f}")
    return history

  # Evaluation utilities
def evaluate_model(model, loader, criterion, name="Model"):
    model.eval()
    preds, labels = [], []
    total_loss=0
    with torch.no_grad():
        for x,y in loader:
            out=model(x)
            total_loss+=criterion(out,y).item()
            preds.extend(out.argmax(1).numpy())
            labels.extend(y.numpy())
    avg_loss = total_loss/len(loader)
    acc = np.mean(np.array(preds)==np.array(labels))
    print(f"\n{name} Test Accuracy: {acc:.4f}")
    print(classification_report(labels, preds, digits=4))
    cm = confusion_matrix(labels,preds)
    sns.heatmap(cm,annot=True,fmt="d"); plt.title(f"{name} Confusion Matrix"); plt.show()
    return acc, cm, labels, preds

def plot_history(hist, title):
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1); plt.plot(hist['train_loss'],label='train'); plt.plot(hist['val_loss'],label='val'); plt.title(title+" Loss"); plt.legend()
    plt.subplot(1,2,2); plt.plot(hist['train_acc'],label='train'); plt.plot(hist['val_acc'],label='val'); plt.title(title+" Accuracy"); plt.legend()
    plt.show()

# Train Custom CNN with Adam, SGD, SGD+Momentum
criterion = nn.CrossEntropyLoss()
optimizers_cfg = {
    'adam': lambda m: optim.Adam(m.parameters(), lr=1e-3),
    'sgd':  lambda m: optim.SGD(m.parameters(), lr=0.01),
    'sgd_mom': lambda m: optim.SGD(m.parameters(), lr=0.01, momentum=0.9)
}

results=[]
for name,opt_fn in optimizers_cfg.items():
    print(f"\n===== Training with {name.upper()} =====")
    model = SimpleCNN()
    optimizer = opt_fn(model)
    hist = train_model(model, optimizer, criterion, train_loader, val_loader, epochs=20)
    plot_history(hist, f"Custom CNN ({name})")
    acc,cm,ytrue,ypred = evaluate_model(model, test_loader, criterion, name=f"Custom CNN ({name})")
    results.append({'optimizer':name,'accuracy':acc})

print("\nOptimizer comparison summary:")
print(pd.DataFrame(results))

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report


# Data preparation
resize_tf = transforms.Compose([
    transforms.Resize((96, 96)),
    transforms.Grayscale(num_output_channels=3),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

mnist_train_tl = datasets.MNIST(root="./data", train=True, download=True, transform=resize_tf)
mnist_test_tl = datasets.MNIST(root="./data", train=False, download=True, transform=resize_tf)

full_tl = torch.utils.data.ConcatDataset([mnist_train_tl, mnist_test_tl])
subset_size = 10000
subset_tl, _ = random_split(full_tl, [subset_size, len(full_tl) - subset_size])

n_train = int(0.70 * subset_size)
n_val = int(0.15 * subset_size)
n_test = subset_size - n_train - n_val

train_tl, val_tl, test_tl = random_split(subset_tl, [n_train, n_val, n_test])

train_loader_tl = DataLoader(train_tl, batch_size=64, shuffle=True)
val_loader_tl = DataLoader(val_tl, batch_size=64, shuffle=False)
test_loader_tl = DataLoader(test_tl, batch_size=64, shuffle=False)

print(f"Train: {n_train}, Val: {n_val}, Test: {n_test}")


def build_finetune_model(base="vgg16"):
    """Build a fine-tuned model based on VGG16 or ResNet50"""
    if base == "vgg16":
        net = models.vgg16(weights="IMAGENET1K_V1")
        for param in net.features.parameters():
            param.requires_grad = False
        net.classifier[-1] = nn.Linear(4096, 10)
    elif base == "resnet50":
        net = models.resnet50(weights="IMAGENET1K_V1")
        for param in list(net.parameters())[:-20]:
            param.requires_grad = False
        net.fc = nn.Linear(net.fc.in_features, 10)

    return net


def train_tl_model(model, train_loader, val_loader, epochs=5):
    """Train transfer learning model"""
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-4)
    criterion = nn.CrossEntropyLoss()

    train_accs = []
    val_accs = []

    for ep in range(epochs):
        # Training
        model.train()
        running_loss = 0
        correct = 0
        total = 0

        for x, y in train_loader:
            optimizer.zero_grad()
            out = model(x)
            loss = criterion(out, y)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
            total += y.size(0)

        train_acc = correct / total
        train_accs.append(train_acc)

        # Validation
        model.eval()
        v_correct = 0
        v_total = 0
        v_loss = 0

        with torch.no_grad():
            for x, y in val_loader:
                out = model(x)
                v_loss += criterion(out, y).item()
                v_correct += (out.argmax(1) == y).sum().item()
                v_total += y.size(0)

        val_acc = v_correct / v_total
        val_accs.append(val_acc)

        print(f"Epoch {ep+1}/{epochs}: train_acc={train_acc:.4f} val_acc={val_acc:.4f} train_loss={running_loss/len(train_loader):.4f}")

    return model, train_accs, val_accs


def evaluate_model(model, test_loader, criterion, name="Model"):
    """Evaluate model and return metrics"""
    model.eval()
    correct = 0
    total = 0
    test_loss = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for x, y in test_loader:
            out = model(x)
            test_loss += criterion(out, y).item()
            preds = out.argmax(1)
            correct += (preds == y).sum().item()
            total += y.size(0)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(y.cpu().numpy())

    accuracy = correct / total
    cm = confusion_matrix(all_labels, all_preds)

    print(f"\n{name} Test Accuracy: {accuracy:.4f}")
    print(f"Test Loss: {test_loss/len(test_loader):.4f}")

    return accuracy, cm, all_preds, all_labels


def plot_confusion_matrix(cm, title="Confusion Matrix"):
    """Plot confusion matrix"""
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=range(10), yticklabels=range(10))
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()


# Main training loop
results = []

for base in ["vgg16", "resnet50"]:
    print(f"\n{'='*60}")
    print(f"Fine-tuning {base.upper()}")
    print('='*60)

    model_tl = build_finetune_model(base)
    model_tl, train_accs, val_accs = train_tl_model(
        model_tl, train_loader_tl, val_loader_tl, epochs=5
    )

    acc, cm, preds, labels = evaluate_model(
        model_tl, test_loader_tl, nn.CrossEntropyLoss(), name=f"{base.upper()}"
    )

    results.append({
        'model': base.upper(),
        'accuracy': acc,
        'train_accs': train_accs,
        'val_accs': val_accs
    })

    # Plot confusion matrix
    plot_confusion_matrix(cm, title=f"{base.upper()} Confusion Matrix")

    # Print classification report
    print(f"\nClassification Report for {base.upper()}:")
    print(classification_report(labels, preds, target_names=[str(i) for i in range(10)]))

# Final summary
print("\n" + "="*60)
print("FINAL MODEL COMPARISON")
print("="*60)

df = pd.DataFrame([{'model': r['model'], 'accuracy': r['accuracy']} for r in results])
print(df)

# Plot accuracy comparison
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='model', y='accuracy', palette='viridis')
plt.title("Transfer Learning Model Test Accuracies on MNIST", fontsize=14, fontweight='bold')
plt.ylabel("Test Accuracy", fontsize=12)
plt.xlabel("Model", fontsize=12)
plt.ylim(0, 1)
for i, row in df.iterrows():
    plt.text(i, row['accuracy'] + 0.01, f"{row['accuracy']:.4f}",
             ha='center', va='bottom', fontsize=11, fontweight='bold')
plt.tight_layout()
plt.show()

# Plot training curves
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
for i, result in enumerate(results):
    axes[i].plot(result['train_accs'], label='Train Accuracy', marker='o')
    axes[i].plot(result['val_accs'], label='Validation Accuracy', marker='s')
    axes[i].set_title(f"{result['model']} Training Curves")
    axes[i].set_xlabel('Epoch')
    axes[i].set_ylabel('Accuracy')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
